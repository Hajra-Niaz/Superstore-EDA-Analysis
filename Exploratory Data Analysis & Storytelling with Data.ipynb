# STEP 2: Clean the Data

```python
#Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
#Check column names
df.columns
```

```python
#Then Dates(string into real format datetime):
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
#Create Useful Columns(for graphs)
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

# STEP 3: Start EDA (For Find Insights)

```python
#Sales by Category
df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

```python
#Profit by Region
df.groupby("Region")["Profit"].sum().sort_values(ascending=False)
```

```python
#Correlation
df[['Sales','Profit','Discount','Quantity']].corr()
```

# STEP 4: Create Visualizations

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1: Bar Chart – Sales by 

Explanation:

This visualization compares total sales across product categories. The Technology category generates the highest revenue, indicating strong customer demand, while Office Supplies contributes the least to overall sales performance.

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Category", y="Sales", data=df)
plt.title("Sales by Category")
plt.show()
```

# 2: Line Chart – Sales Over Time

Explanation:

The line graph illustrates monthly sales trends, showing fluctuations throughout the year with a noticeable peak in the later months. This suggests possible seasonal demand or increased purchasing activity toward year-end.

```python
monthly_sales = df.groupby("Month")["Sales"].sum()

plt.figure(figsize=(8,4))
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# 3: Histogram – Profit Distribution

Explanation:

The histogram highlights the distribution of profit across transactions. While most orders generate small positive profits, there are significant negative outliers, indicating that certain transactions result in substantial losses.

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Profit"], bins=30)
plt.title("Profit Distribution")
plt.show()
```

# 4: Heatmap – Correlation

Explanation:

The correlation heatmap reveals a moderate positive relationship between Sales and Profit, indicating that higher sales generally contribute to increased profitability. However, there is a negative correlation between Discount and Profit, suggesting that excessive discounting reduces overall profit margins.

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Sales','Profit','Discount','Quantity']].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# 5: Scatter Plot – Discount vs Profit

Explanation:

The scatter plot demonstrates a clear negative relationship between discount and profit. As discount levels increase, profits tend to decline, with several high-discount transactions resulting in losses. This highlights the financial impact of aggressive discount strategies.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Discount", y="Profit", data=df)
plt.title("Discount vs Profit")
plt.show()
```
